```python
Y=data_train['price_range']
```

```python
Y
```

Scaling datset

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaler = StandardScaler()
scaler.fit(X)
X_transformed = scaler.transform(X)
X_train,X_test,Y_train,Y_test = train_test_split(X_transformed,Y,test_size=0.3)
```

# Building primary model 

Building one model using random forest classifier

```python
from sklearn.metrics import f1_score, accuracy_score, recall_score, roc_auc_score, confusion_matrix, classification_report, plot_confusion_matrix, precision_score, ConfusionMatrixDisplay
from  sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
```

```python
random_forest = RandomForestClassifier(n_estimators=10, criterion = 'entropy', random_state=101)
random_forest.fit(X_train, Y_train)
predRFC = random_forest.predict(X_test)
reportRFC = classification_report(Y_test,predRFC,output_dict = True)
pd.DataFrame(reportRFC).transpose()
```

```python
random_forest.score(X_test,Y_test)
```

Building one model using logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report


model = LogisticRegression()
model.fit(X_train,Y_train)
Y_train_pred = model.predict(X_train)
Y_test_pred = model.predict(X_test)


print("Train Set Accuracy:"+str(accuracy_score(Y_train_pred,Y_train)*100))
print("Test Set Accuracy:"+str(accuracy_score(Y_test_pred,Y_test)*100))
print("\nConfusion Matrix:\n%s"%confusion_matrix(Y_test_pred,Y_test))
print("\nClassification Report:\n%s"%classification_report(Y_test_pred,Y_test))
```

```python
model.score(X_test,Y_test)
```

# Cross-validation and Hyper-parameter Tuning

Identifying various hyper-parameters and performing randomizedsearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
import numpy as np
```

```python
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
random_grid = {'n_estimators': [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)],
               'max_features': ['auto', 'sqrt'],
               'max_depth': max_depth,
               'min_samples_split': [2, 5, 10],
               'min_samples_leaf': [1, 2, 4],
               'bootstrap': [True, False]}

clf = RandomForestClassifier()
modelrf = RandomizedSearchCV(estimator = clf, param_distributions = random_grid, n_iter = 100, cv = 4, verbose=2, random_state=42, n_jobs = -1)
modelrf.fit(X_train, Y_train)
modelrf.best_params_

# this prints the contents of the parameters in the random grid
print ('Random grid: ', random_grid, '\n')

# print the best parameters
print ('Best Parameters: ', modelrf.best_params_, ' \n')
```

```python
Y_pred_randomized = modelrf.predict(X_test)
print(classification_report(Y_pred_randomized, Y_test))
```

```python
modelrf.score(X_test,Y_test)
```
